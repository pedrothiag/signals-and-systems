## Plotando Sinais de Tempo Contínuo

```python
import numpy as np
import matplotlib.pyplot as plt

t_inicio = 0
t_fim = 1
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)
x = np.cos(2*np.pi*10*t)  #x(t) = cos(2*pi*10*t)

plt.figure(figsize=(7,3))
plt.plot(t,x, 'k')
plt.xlabel('$t$')
plt.ylabel('$x(t) = \\cos(2\\pi\\cdot 10t)$')
plt.xlim([t_inicio, t_fim])
plt.grid()
plt.show()
```

Vamos plotar o degrau unitário de tempo contínuo. Relembrando:
$$
u(t) = \begin{cases}
1 & t \geq 0 \\
0 & \mathrm{c.c.}
\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

t_inicio = -2
t_fim = 5
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints) 

x = 1.0*(t>=0)

plt.figure(figsize=(7,3))
plt.plot(t,x)
plt.xlabel('$t$')
plt.ylabel('$u(t)$')
plt.xlim([t_inicio, t_fim])
plt.grid()
plt.show()
```

Vamos gerar diferentes sinais exponenciais reais, ou seja:
$$
x(t) = e^{at}
$$
considerando: (a) $a= 1,2$, (b) $a = -1,2$. Vamos verificar o comportamento de crescimento ou decaimento para cada caso.

```python
t_inicio = -3
t_fim = 4
npoints = 1000

t = np.linspace(t_inicio, t_fim, npoints)

x1 = np.exp(1.2*t)
x2 = np.exp(-1.2*t)

plt.figure()
plt.subplot(2,1,1)
plt.plot(t,x1, 'r')
plt.xlabel('$t$')
plt.ylabel('$x_1(t)$')
plt.title('$a > 0$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,1,2)
plt.plot(t,x2, 'k')
plt.xlabel('$t$')
plt.ylabel('$x_2(t)$')
plt.title('$a < 0$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.tight_layout()
plt.show()
```

```python
u = 1.0*(t>=0)

x3 = np.exp(-1.2*t)*u
plt.plot(t,x3, 'k')
plt.xlabel('$t$')
plt.ylabel('$x_3(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])
plt.show()
```

## Operações com Sinais

### Operações Envolvendo a Amplitude

Dado os seguintes sinais:
$$
x_1(t) = \sin (10\pi t)
$$
$$
x_2(t) = u(t)
$$
$$
x_3(t) = e^{-2t}u(t)
$$
Determine os seguintes sinais: (a) $x_1(t) + x_3(t)$, (b) $2x_1(t)$, (c) $x_1(t)x_2(t)$ e (d) $1,5x_1(t) + x_2(t) + 2,5x_3(t)$

```python
t_inicio = -2
t_fim = 3
npoints = 5000
t = np.linspace(t_inicio, t_fim, npoints)

x1 = np.sin(10*np.pi*t)
x2 = 1.0*(t>=0)
x3 = np.exp(-2*t)*(t>=0)

xadd = x1 + x3
xgain = 2*x1
xmul = x1*x2
xcomb = 1.5*x1 + x2 + 2.5*x3

plt.figure()
plt.subplot(3,1,1)
plt.plot(t,x1)
plt.xlabel('$t$')
plt.ylabel('$x_1(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(3,1,2)
plt.plot(t,x2)
plt.xlabel('$t$')
plt.ylabel('$x_2(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(3,1,3)
plt.plot(t,x3)
plt.xlabel('$t$')
plt.ylabel('$x_3(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(t,xadd)
plt.xlabel('$t$')
plt.ylabel('$x_1(t) + x_3(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,2,2)
plt.plot(t,xgain)
plt.xlabel('$t$')
plt.ylabel('$2x_1(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,2,3)
plt.plot(t,xmul)
plt.xlabel('$t$')
plt.ylabel('$x_1(t)x_2(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,2,4)
plt.plot(t,xcomb)
plt.xlabel('$t$')
plt.ylabel('$1.5x_1(t) + x_2(t) + 2.5x_3(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.tight_layout()
plt.show()
```

### Operações Envolvendo o Tempo

Dado o seguinte sinal:

$$
x(t) = \begin{cases}
t + 1 & -1 \leq t \leq 0 \\
-0,5t + 1 & 0 < t \leq 2 \\
0 & \mathrm{c.c.}
\end{cases}
$$

Determine: (a) $x(t/2)$, (b) $x(2t)$, (c) $x(-t)$, (d) $x(t-2)$

```python
def meusinal(t):
    # if (t>=-1) and (t<=0):
    #     return t + 1
    # elif (t>0) and (t<=2):
    #     return -0.5*t + 1
    # else:
    #     return 0
    return (t+1)*(np.logical_and(t>=-1,t<=0)) + (-0.5*t+1)*(np.logical_and(t>0,t<=2))

t_inicio = -6
t_fim = 6
npoints = 1000

t = np.linspace(t_inicio, t_fim, npoints)
x = meusinal(t)
# for idx in range(len(t)):
#     x[idx] = meusinal(t[idx])

plt.figure()
plt.plot(t, x)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])
```

```python
x1 = meusinal(t/2)
x2 = meusinal(2*t)
x3 = meusinal(-t)
x4 = meusinal(t-2)

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(t,x1)
plt.xlabel('$t$')
plt.title('$x(t/2)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,2,2)
plt.plot(t,x2)
plt.xlabel('$t$')
plt.title('$x(2t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,2,3)
plt.plot(t,x3)
plt.xlabel('$t$')
plt.title('$x(-t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,2,4)
plt.plot(t,x4)
plt.xlabel('$t$')
plt.title('$x(t-2)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.tight_layout()
plt.show()
```

## Energia de Sinais

Determine, computacionamente, a energia do seguinte sinal de tempo contínuo,

$$
x(t) = u(t) - u(t-1)
$$
 
Compare com o resultado analítico $E_x = 1$.

```python
t_inicio = -2
t_fim = 5
npoints = 1000

t = np.linspace(t_inicio, t_fim, npoints)
dt = (t_fim - t_inicio)/npoints

u = 1.0*(t>=0)
u1 = 1.0*(t>=1)
x = u - u1

plt.figure(figsize=(7,3))
plt.plot(t,x)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])
plt.show()

Ex = dt*np.sum(x**2)   #Aproximação para a Energia
print('Energia (computacional) = ', Ex)
print('Energia (analítica) = ',1)
```

## Sistemas de Tempo Contínuo

Implemente um sistema modulador em amplitude, definido pela seguinte relação entrada-saída:

$$
y(t) = x(t)\cos(2\pi F_c t)
$$

sendo $F_c$ a frequência da portadora e $x(t)$ o sinal de entrada. Simule considerando $F_c = 10$ kHz e $x(t) = \cos(2\pi F_m t)$, sendo $F_m = 500$ Hz.

```python
t_inicio = 0
t_fim = 0.01
npoints = 10_000

t = np.linspace(t_inicio, t_fim, npoints)

Fm = 500
x = np.cos(2*np.pi*Fm*t)

Fc = 10_000
y = x*np.cos(2*np.pi*Fc*t)

plt.figure(figsize=(7,4))
plt.subplot(2,1,1)
plt.plot(t,x)
plt.xlabel('$t$')
plt.title('$x(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,1,2)
plt.plot(t,y)
plt.xlabel('$t$')
plt.title('$y(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(t,x, 'r', label='$x(t)$')
plt.plot(t,y, label='$y(t)$', alpha=0.5)
plt.xlabel('$t$')
plt.grid()
plt.xlim([t_inicio, t_fim])
plt.legend(loc='upper right')
plt.show()
```

```python
def modulador(x):
    Fc = 10_000
    return x*np.cos(2*np.pi*Fc*t)
```

```python
t_inicio = 0
t_fim = 0.01
npoints = 10_000

t = np.linspace(t_inicio, t_fim, npoints)
Fm = 500
x = np.cos(2*np.pi*Fm*t)

y = modulador(x)

plt.figure(figsize=(10,5))
plt.plot(t,x, 'r', label='$x(t)$')
plt.plot(t,y, label='$y(t)$', alpha=0.5)
plt.xlabel('$t$')
plt.grid()
plt.xlim([t_inicio, t_fim])
plt.legend(loc='upper right')
plt.show()
```

## Sistemas Lineares e Invariantes no Tempo

Determine a saída de um sistema LTI com resposta ao impulso:
$$
h(t) = \begin{cases}
t, & 0 < t < 2 \\
0, & \mathrm{c.c.}
\end{cases}
$$
em que a entrada é:
$$
x(t) = \begin{cases}
1, & 0 < t < 1 \\
0, & \mathrm{c.c.}
\end{cases}
$$

```python
t_inicio = -2
t_fim = 4
npoints = 10_000

t = np.linspace(t_inicio, t_fim, npoints)
x = 1.0*(np.logical_and(t>0, t<1))
h = t*(np.logical_and(t>0, t<2))

plt.figure()
plt.subplot(2,1,1)
plt.plot(t, x)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,1,2)
plt.plot(t, h)
plt.xlabel('$t$')
plt.ylabel('$h(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.tight_layout()
plt.show()
```

```python
print(x.shape)
print(h.shape)
```

```python
dt = (t_fim - t_inicio)/npoints
y = dt*np.convolve(x,h)
print(y.shape)
```

```python
t_inicio_y = 2*t_inicio
t_fim_y = 2*t_fim

t_y = np.linspace(t_inicio_y, t_fim_y, len(y))

plt.figure()
plt.plot(t_y, y)
plt.xlabel('$t$')
plt.ylabel('$y(t) = x(t) \\ast h(t)$')
plt.grid()
plt.xlim([t_inicio_y, t_fim_y])
plt.show()
```

Quando temos sinais de duração infinita, a convolução computacional será igual à convolução real somente no intervalo de tempo dos sinais originais. Por exemplo, determine a saída de um sistema LTI com resposta ao impulso:
$$
h(t) = e^{-2t}u(t)
$$
em que a entrada é:
$$
x(t) = u(t)
$$

```python
t_inicio = -2
t_fim = 4
npoints = 10_000

t = np.linspace(t_inicio, t_fim, npoints)

h = np.exp(-2*t)*(t>=0)
x = 1.0*(t>=0)

plt.figure(figsize=(7,4))
plt.subplot(2,1,1)
plt.plot(t, x)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.subplot(2,1,2)
plt.plot(t, h)
plt.xlabel('$t$')
plt.ylabel('$h(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])

plt.show()
```

```python
dt = (t_fim - t_inicio)/npoints

y = dt*np.convolve(x,h)

t_inicio_y = 2*t_inicio
t_fim_y = 2*t_fim
t_y = np.linspace(t_inicio_y, t_fim_y, len(y))

plt.figure(figsize=(7,3))
plt.plot(t_y, y)
plt.xlabel('$t$')
plt.ylabel('$h(t)$')
plt.grid()
plt.xlim([t_inicio, t_fim])
plt.show()
```

## Equações Diferenciais Ordinárias

Vamos considerar um sistema de tempo contínuo descrito pela seguinte equação diferencial:

$$
\frac{d^2y(t)}{dt^2} + 5\frac{dy(t)}{dt} + 6y(t) = 6x(t)
$$

Inicialmente devemos criar a equação diferencial. Para simplificar, vamos primeiro determinar a resposta natural dessa equação, ou seja, considerando $x(t) = 0$ temos:

$$
\frac{d^2y(t)}{dt^2} + 5\frac{dy(t)}{dt} + 6y(t) = 0
$$

```python
import sympy as sp
#sp.init_printing()       #Melhora a apresentacao das ED

t = sp.symbols('t')       #Cria a variavel simbolica t
y = sp.Function('y')      #Cria uma funcao y

eq = sp.diff(y(t),t,t) + 5*sp.diff(y(t),t) + 6*y(t)  #Cria a ED. Quando usa t,t é a derivada segunda
eq
```

```python
sp.dsolve(eq)
```

Verifique que nesse caso, a resposta natural ainda está em função das constantes arbitárias. Vamos complicar um pouco mais e adicionar uma entrada, no caso $x(t) = \frac{2}{3}e^{-t}$. Se substituirmos $x(t)$ na equação diferencial vemos que:


$$
\frac{d^2y(t)}{dt^2} + 5\frac{dy(t)}{dt} + 6y(t) = 6\left(\frac{2}{3}e^{-t}\right)
$$

o que é a mesma coisa de:

$$
\frac{d^2y(t)}{dt^2} + 5\frac{dy(t)}{dt} + 6y(t) - 4e^{-t} = 0
$$

É essa equação que deveremos implementar.

```python
eq = sp.diff(y(t),t,t) + 5*sp.diff(y(t),t) + 6*y(t) - 4*sp.exp(-t)
eq
```

```python
sp.dsolve(eq)
```

Ao resolvermos, note que a resposta já engobla a resposta forçada + resposta natural. Para resolver adicionando condições iniciais, podemos complementar o comando através da diretiva `ics`. Se quisermos calcular a resposta considerando as condições iniciais $y(0) = 10$ e $\dot{y}(0) = -2$, poderemos fazer:

```python
initial_condition = {
    y(t).subs(t,0):10, 
    y(t).diff(t).subs(t, 0):-2
}
sp.dsolve(eq, ics=initial_condition)
```

Verifique que essa é a mesma resposta quando utilizamos o método clássico! Em resumo, o código abaixo resolve a seguinte equação diferencial:

$$
\frac{d^2y(t)}{dt^2} + 5\frac{dy(t)}{dt} + 6y(t) = 6x(t)
$$

com entrada $x(t) = \frac{2}{3}e^{-t}$ e condições iniciais $y(0) = 10$ e $\dot{y}(0) = -2$.

```python
import sympy as sp

t = sp.symbols('t')   #Cria a variavel simbolica t
y = sp.Function('y')  #Cria uma funcao y
x = sp.Function('x')

x = 2/3*sp.exp(-t)

eq = sp.diff(y(t),t,t) + 5*sp.diff(y(t),t) + 6*y(t) - 6*x

initial_condition = {
    y(t).subs(t,0):10, 
    y(t).diff(t).subs(t, 0):-2
}
sol = sp.dsolve(eq, ics=initial_condition)
sol
```

A resposta ao impulso de um sistema pode ser determinada de forma semelhante, utilizando a método de casamento de impulso. Por exemplo, determine a resposta ao impulso do sistema descrito pela seguinte equação diferencial:

$$
\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = \frac{dx(t)}{dt}
$$

```python
t = sp.symbols('t')   #Cria a variavel simbolica t
y = sp.Function('y')  #Cria uma funcao y

eq = sp.diff(y(t),t,t) + 3*sp.diff(y(t),t) + 2*y(t)   #Desejamos apenas a componente de entrada nula

initial_conditions = {
    y(t).subs(t,0):0, 
    y(t).diff(t).subs(t, 0):1
}
yh = sp.dsolve(eq,ics=initial_conditions)
yh
```

```python
h = sp.diff(yh.rhs,t)
h
```
